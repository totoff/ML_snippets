```
from transformers import GPT2TokenizerFast
```

```
tokenizer = GPT2TokenizerFast.from_pretrained('gpt2')
```

```
tokenizer.encode('i was very happy')
```

```
from transformers import OpenAIGPTLMHeadModel, OpenAIGPTTokenizer
tokenizer = OpenAIGPTTokenizer.from_pretrained('openai-gpt')
model = OpenAIGPTLMHeadModel.from_pretrained('openai-gpt')
#generate 4 sequences started with i was very  with GPT
generated_sequences = model.generate(input_ids=tokenizer.encode('i was very', return_tensors='pt'),
                                     max_length=50,
                                     do_sample=True,
                                     num_beams=1,
                                     temperature=0.7,
                                     top_p=0.8,
                                     eos_token_ids=tokenizer.unk_token_id,
                                     repetition_penalty=1.4,
                                     num_return_sequences=4)
for sequence in generated_sequences[0]:
  print('<>' + tokenizer.decode(sequence))
```
