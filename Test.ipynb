```python
from pandas import read_csv
from matplotlib import pyplot
# load the dataset
data = read_csv('daily-total-female-births.csv')
```

```python
import pandas as pd
X_ans=[]
Y_ans=[]
for i in range(len(data["Births"])-3):
    X = list(data["Births"])[i:i+3]
    Y = list(data["Births"])[i+3]
    X_ans.append(X)
    Y_ans.append(Y)
print(Y_ans)
print(X_ans)
```

```python
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
# define dataset
X_birth = array(X_ans)
y_birth = array(Y_ans)
last_id = int(y_birth.shape[0]*0.8)
X_train = X_birth[:last_id]
y_train = y_birth [:last_id]
X_test = X_birth[last_id:]
y_test = y_birth[last_id:]
# define model
model = Sequential()
model.add(Dense(100, activation='relu', input_dim=3))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
# fit model
model.fit(X_train, y_train, epochs=2000, verbose=0)
# demonstrate prediction
yhat = model.predict(X_test, verbose=0)
print(yhat)
print(y_test)
```

```python
# univariate cnn example
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
# define dataset
X_birth = array(X_ans)
y_birth = array(Y_ans)
last_id = int(y_birth.shape[0]*0.8)
X_train = X_birth[:last_id]
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
y_train = y_birth [:last_id]
X_test = X_birth[last_id:]
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
y_test = y_birth[last_id:]
# define model
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(3, 1)))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(50, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
# fit model
model.fit(X_train, y_train, epochs=1000, verbose=0)
# demonstrate prediction
yhat = model.predict(X_test, verbose=0)
print(yhat)
print(y_test)
```

```python
import pandas as pd
X_ans=[]
Y_ans=[]
for i in range(len(data["Births"])-4):
    X = list(data["Births"])[i:i+4]
    Y = list(data["Births"])[i+4]
    X_ans.append(X)
    Y_ans.append(Y)
```

```python
from numpy import array
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import TimeDistributed
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
# define dataset
X_birth = array(X_ans)
y_birth = array(Y_ans)
last_id = int(y_birth.shape[0]*0.8)
X_train = X_birth[:last_id]
# reshape from [samples, timesteps] into [samples, subsequences, timesteps, features]
X_train = X_train.reshape((X_train.shape[0], 2, 2, 1))
y_train = y_birth [:last_id]
X_test = X_birth[last_id:]
# reshape from [samples, timesteps] into [samples, subsequences, timesteps, features]
X_test = X_test.reshape((X_test.shape[0], 2, 2, 1))
y_test = y_birth[last_id:]
# define model
model = Sequential()
model.add(TimeDistributed(Conv1D(filters=64, kernel_size=1, activation='relu'), input_shape=(None, 2, 1)))
model.add(TimeDistributed(MaxPooling1D(pool_size=2)))
model.add(TimeDistributed(Flatten()))
model.add(LSTM(50, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
# fit model
model.fit(X_train, y_train, epochs=500, verbose=0)
yhat = model.predict(X_test, verbose=0)
print(yhat)
print(y_test)
```
